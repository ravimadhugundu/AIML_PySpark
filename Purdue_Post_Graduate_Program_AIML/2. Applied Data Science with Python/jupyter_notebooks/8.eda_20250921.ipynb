# EDA
- EDA stands for Exploratory Data Analysis
- It's a method of analyzing and summarizing data sets before making any assumptions or sometimes performing data cleansing.
- Advantages: 
    - identify patterns - getting familiar with the data
    - spot anomalies
    - data quality inspection
    - testing hypotheses
- Identifying data issues. During EDA, you may encounter issues like inconsistencies, duplicate records, or irrelevant features. Detecting these early on saves time and resources, ensuring you're not working with faulty data when building models.
- Defining the project objective is essential:
    - What problems are we trying to solve?
    - What are the KPIs (key performance indicators) for success?
    - What insights or outcomes do stakeholders expect?
    - What data elements I need or have accessible to solve for the problem?
    - The proper solution for the problem:
        - Statistical Analysis
        - Traditional Machine Learning
        - Advanced techniques like Deep Learning
        - Dashboard
    - If applicable: legal, privacy, and ethical considerations
- EDA Techniques
    - **Descriptive Statistical Measures**
        - Measures of Central Tendency: e.g. mean, mode, etc..
        - Measures of Dispersion: e.g. Stand Dev
        - Measures of Shape: e.g. Skewness
        - Measures of Relationship: e.g. correlation
    - **Non-Graphical Analysis**
        - Data quality checks:
            - missing values
            - appropriate data types
            - outliers
            - duplicates
        - Aggregation: e.g.
            - `groupby()`
            - `pivot_table()`
    - **Graphical Analysis** 
        - using data visualization techniques
        - popular Python libraries: `pandas`, `matplotlib`, `seaborn`, `plotly`, and `bokeh`
        - chart examples:
            - Histograms
            - Box and whisker
            - Scatter plots
            - Bar charts
            - Pie charts
            - Line plots

### Line Plots

```python
# generate data 
x = np.linspace(0, 10, 100)
y = np.sin(x)
```

```python
#create a line plot 
plt.plot(x, y, label='Sine Wave', color='green', linewidth=4)

#formatting to the plot area
plt.xlabel('x-axis', fontsize=14, color='purple')
plt.ylabel('y-axis', fontsize=14, color='purple')
plt.title('Line Plot of A Sine Wave', fontsize=20, color='red')
plt.legend()

#add grid
plt.grid(True, linestyle='--', alpha=0.7) #other styles '-', '.', '-.'
```

Multiple elements/datasets in one plot

```python
# generate data 
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
#change the size of the plot
plt.figure(figsize=(12,6))

#create a line plot 
plt.plot(x, y1, label='Sine Wave', color='#178b58', linewidth=3)
plt.plot(x, y2, label='Cosine Wave', color="#307fd9", linewidth=3, linestyle='--')

#formatting to the plot area
plt.xlabel('x-axis', fontsize=14, color='purple')
plt.ylabel('y-axis', fontsize=14, color='purple')
plt.title('Line Plot of Sine vs Cosine', fontsize=20, color='red')
plt.legend(loc='upper center')
```

```python
my_plot_cols = ['SalePrice', 'YearBuilt', 'LotArea', '1stFlrSF', 'GarageArea','GrLivArea']

fig, axes = plt.subplots(nrows=2,ncols=3, figsize=(16,6), dpi=300)

#flattening the grid axes
axes = axes.flatten() #convert to 0,1,2,3,... locations

for idx, col in enumerate(my_plot_cols):
    if col == 'LotArea':
        sns.boxplot(data=df, x=col, ax=axes[idx])    
    else:
        sns.histplot(data=df, x=col, ax=axes[idx])

plt.tight_layout()
```

```python
end_point = df['Neighborhood'].value_counts().max() + 20
step = end_point//20 #double slash will use non-decimal value

plt.figure(figsize=(16,8))
ax = sns.countplot(data=df, y='Neighborhood'
              , order=df['Neighborhood'].value_counts().index   
              #, hue='Neighborhood'  
              #, palette='tab20b'   
                    )

plt.xticks(range(0,end_point,10))
plt.title('Neighborhood by Sales Volume')

;
```

### Multi-variate Analysis

- It allows for relationship or association analysis
- Types:
    - Numerical vs Numerical 
    - Numerical vs Categorical
    - Categorical vs Categorical 

```python
plt.figure(figsize=(7,7))

sns.scatterplot(x=df['SalePrice'], y=df['GrLivArea'])
plt.ylabel('Above grade (ground) living area square feet')
```

- We see a positive correlation (aside from some of the outliers)
- To dive further into other factors, we can expand the scatter plot to a bubble chart

```python
plt.figure(figsize=(16,7))

sns.scatterplot(x=df['SalePrice'], y=df['GrLivArea'],alpha=0.6,
                size=df['YearBuilt'],
                hue=df['BsmtQual'],
                style=df['Foundation'])
plt.ylabel('Above grade (ground) living area square feet')
```

- The higher the price, the higher the basement quality is for the houses

`regplot`

```python
plt.figure(figsize=(15,7), dpi=300)

sns.regplot(x=df['SalePrice'], y=df['GrLivArea'], line_kws={'color':'navy'})
plt.ylabel('Above grade (ground) living area square feet')
```

`jointplot` offer multiple features in one view:
- Scatter plot
- Regression fitted line
- Histogram

```python
plt.figure(figsize=(15,7), dpi=300)

sns.jointplot(data=df, x='SalePrice', y='GrLivArea', kind='reg', line_kws={'color':'navy'},
              scatter_kws = {'color':'green', 'alpha':0.6, 's':4})
plt.ylabel('Above grade (ground) living area square feet')
```
